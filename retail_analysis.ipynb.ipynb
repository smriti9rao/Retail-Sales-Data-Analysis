### Visualization using Plotly

```python
plt.figure(figsize=(7, 4))
sns.histplot(data=df, x='price', hue='category', multiple="stack", kde=False)
plt.title('Price Distribution by Shopping Mall and Category')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(data=df, x='quantity', hue='gender', multiple="stack", kde=False)

plt.title('Quantity Distribution by Category and Gender')
plt.xlabel('Quantity')
plt.ylabel('Frequency')
plt.show()
```

```python
payment_method_counts = df['payment_method'].value_counts()
plt.figure(figsize=(4, 4))
plt.pie(payment_method_counts, labels=payment_method_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Payment Methods')
plt.axis('equal') 
plt.show()
```

```python

plt.figure(figsize=(7, 3))
sns.histplot(data=df, x='price', hue='category', kde=False, bins=10)
plt.title('Price Distribution by Category')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python

df['invoice_date'] = pd.to_datetime(df['invoice_date'], errors='coerce', dayfirst=True)

df['month_year'] = df['invoice_date'].dt.to_period('M').astype(str)

df['total_price'] = df['quantity'] * df['price']

monthly_sales = df.groupby('month_year')['total_price'].sum().reset_index()
plt.figure(figsize=(7, 3))
plt.plot(monthly_sales['month_year'], monthly_sales['total_price'], marker='o')

plt.title("Sales Trend Over Time")
plt.xlabel('Month-Year')
plt.ylabel('Total Sales Revenue')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

```
